### Descarga de datos GSM8K

```python
!pip install -q datasets

from datasets import load_dataset

# Carga GSM8K desde Hugging Face como dataset público
dataset = load_dataset("openai/gsm8k", "main")

print(dataset)
print(dataset["train"][0])
```

```python
import pandas as pd
import random

# Mostrar primeras 10 filas del conjunto de entrenamiento
train_df = pd.DataFrame(dataset['train'][:10])
display(train_df[['question', 'answer']])

# Mostrar 5 ejemplos aleatorios del conjunto de prueba
idxs = random.sample(range(len(dataset['test'])), 5)
test_df = pd.DataFrame([dataset['test'][i] for i in idxs])
display(test_df[['question', 'answer']])
```
